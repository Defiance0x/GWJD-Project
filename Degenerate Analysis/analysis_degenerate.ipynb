```python
def Vectors(b):
    # b is a dataframe
    w = np.array(b.time)
    x = np.array(b.bcurrent)
    y = np.array(b.temperature)
    z = np.array(b.resistance)
    return w,x,y,z
```

```python
def delete_words(g):
    g = np.delete(g, [0])
    return g

def numberfy(an_array):
    new_ar = np.array([])
    for i in an_array:
        i = float(i)
        new_ar = np.append(new_ar, i)
    return new_ar


def obtain_arrays(df):
    w_time,x_bcurrent,y_temperature,z_resistance = Vectors(df)
    w_t = delete_words(w_time)
    x_b = delete_words(x_bcurrent)
    y_T = delete_words(y_temperature)
    z_r = delete_words(z_resistance)
    w_t = numberfy(w_t)
    x_b = numberfy(x_b)
    y_T = numberfy(y_T)
    z_r = numberfy(z_r)
    #y_T, z_r, w_t, x_b = numberdeleter(y_T, z_r, w_t, x_b)
    return y_T, z_r, w_t, x_b
```

```python
def individual_plot(T,R,data,lamp,ndf,cdwu):
    title = data + ' Lamp: '+str(lamp)+ ', ndf:' + str(ndf)+", " +cdwu
    MR = R/1000000
    plt.plot(T,MR)
    plt.title(title)
    #plt.legend()
    plt.xlabel('Temperature (K)')
    plt.ylabel('Resistance (MΩ)')
    return plt.show()



individual_plot(T16_3,R16_3,'16_3',60,12,'warm up')
```

```python
def Tchopping(Told,Rold,told,Bold,Tlower,Tupper):
    # limits the range of temperature values
    Tnew = np.array([])
    Rnew = np.array([])
    tnew = np.array([])
    Bnew = np.array([])
    for i, x in np.ndenumerate(Told):
        if x < Tupper and x > Tlower:
            Tnew = np.append(Tnew,x)
            Rnew = np.append(Rnew, Rold[i])
            tnew = np.append(tnew, told[i])
            Bnew = np.append(Bnew, Bold[i])  
    individual_plot(Told,Rold,'old',0,0,'')
    individual_plot(Tnew,Rnew,'new',0,0,'')
    return Tnew, Rnew, tnew, Bnew
```

```python
def delete_dodgy_R(Told,Rold,told,Bold,Tlower,Tupper,Rlower,Rupper):
    #looks withing a certain T range and deletes any R values if out of limits
    Tnew = np.array([])
    Rnew = np.array([])
    tnew = np.array([])
    Bnew = np.array([])
    for i, x in np.ndenumerate(Told):
        if x < Tlower or x > Tupper:
            Tnew = np.append(Tnew, Told[i])
            Rnew = np.append(Rnew, Rold[i])
            tnew = np.append(tnew, told[i])
            Bnew = np.append(Bnew, Bold[i])  
        else:
            if Rold[i] < Rupper and Rold[i] > Rlower:
                Tnew = np.append(Tnew, Told[i])
                Rnew = np.append(Rnew, Rold[i])
                tnew = np.append(tnew, told[i])
                Bnew = np.append(Bnew, Bold[i]) 
    individual_plot(Told,Rold,'old',0,0,'')
    individual_plot(Tnew,Rnew,'new',0,0,'')
    return Tnew, Rnew, tnew, Bnew
    
```
