```python
def generator(day, max_measurements):

  day = str(day)

  for i in range(1, max_measurements+1):
    print('df%s' % (day + '_' + str(i) + " = " + "drive_to_df('')"))  
  for i in range(1, max_measurements+1):
    T = 'T%s' % (day + '_' + str(i))
    R = 'R%s' % (day + '_' + str(i))
    t = 't%s' % (day + '_' + str(i))
    B = 'B%s' % (day + '_' + str(i))
    print(T,", ", R,", ", t,", ", B, "= obtain_arrays(df%s)" % (day + '_' + str(i)))

  for i in range(1, max_measurements+1):
    T = 'T%s' % (day + '_' + str(i))
    R = 'R%s' % (day + '_' + str(i))
    if i % 2 == 0: # if number is even, prints cool down label
      print("plt.plot(", T, ", ",R, ", ","label = 'df%s -- cool down'" % (day + '_' + str(i)),")")
    else: # else if number is odd, prints odd label
      print("plt.plot(",T,", ",R, ", ","label = 'df%s -- warm up'" % (day + '_' + str(i)),")")

  print("plt.legend()")
  print("plt.xlabel('Temperature (K)')")
  print("plt.ylabel('Resistance (Ω)')")
  print("plt.show()")

  return

#generator(day = 17, max_measurements = 20)
```

```python
def Tchopping(Told,Rold,told,Bold,Tlower,Tupper):
    Tnew = np.array([])
    Rnew = np.array([])
    tnew = np.array([])
    Bnew = np.array([])
    for i, x in np.ndenumerate(Told):
        if x < Tupper and x > Tlower:
            Tnew = np.append(Tnew,x)
            Rnew = np.append(Rnew, Rold[i])
            tnew = np.append(tnew, told[i])
            Bnew = np.append(Bnew, Bold[i])  
    return Tnew, Rnew, tnew, Bnew
    
    
Tn1,Rn1,tnew,Bnew = Tchopping(T1,R1,t1,B1,100,175)
plt.scatter(Tn1,Rn1)
plt.show()
```
