# van der pauw checker

```python
pie = np.pi


def VdP(RS, *data):
    R1234, R2341 = data
    pie = np.pi
    eqn = np.exp(-pie*R1234/RS) + np.exp(-pie*R2341/RS) - 1
    return eqn

VdP(1312.38662461, 40, 1000)

```
